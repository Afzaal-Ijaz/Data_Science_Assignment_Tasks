# Subplot 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
x = np.linspace(-10,10,200)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, x)
axs[0, 0].set_title("y = x")

axs[0, 1].plot(x, x**2)
axs[0, 1].set_title("y = x")

axs[1, 0].plot(x, x**3)
axs[1, 0].set_title("y = x")

axs[1, 1].plot(x, np.sin(x))
axs[1, 1].set_title("y = sin(x)")

plt.tight_layout()
plt.show()
```

```python
#2
x = np.linspace(-2*np.pi, 2*np.pi, 400)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(x, np.cos(x), color='green')
ax1.set_title("y = cos(x)")

ax2.plot(x, np.exp(x), color='purple')
ax2.set_title("y = exp(x)")

plt.tight_layout()
plt.show()
```

```python
#3
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Left subplot
axs[0].plot(x, y1)
axs[0].set_title("y = sin(x)")
axs[0].set_xlabel("X-axis")
axs[0].set_ylabel("Y-axis")

# Right subplot
axs[1].plot(x, y2)
axs[1].set_title("y = cos(x)")
axs[1].set_xlabel("X-axis")
axs[1].set_ylabel("Y-axis")

plt.tight_layout()
plt.show()
```

# Filled Area Plot

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y, label='y = sin(x)', color='b')

plt.fill_between(x, y, 0, color='red', alpha=0.5)

plt.title("y = sin(x) with area under curve")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 500)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.fill_between(x, y1, y2, where=(y1 > y2), interpolate=True, color='blue', alpha=0.5, label='sin(x) > cos(x)')
plt.fill_between(x, y1, y2, where=(y1 <= y2), interpolate=True, color='green', alpha=0.5, label='cos(x)  sin(x)')
plt.title("Filled Area Between sin(x) and cos(x)")
plt.legend()
plt.show()
```

# 3D line Plot

```python
fig = plt.figure(figsize=(8, 6))
axs = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
z = x + y

axs.plot(x, y, z, label='z = x + y')
axs.set_title("3D Line Plot")
axs.set_xlabel("x")
axs.set_ylabel("y")
axs.set_zlabel("z")
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

x = np.random.randint(0, 50, 200)
y = np.random.randint(0, 50, 200)
z = np.random.randint(0, 50, 200)

ax.scatter(x, y, z, c='red', marker='o')
ax.set_title("3D Scatter Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```
