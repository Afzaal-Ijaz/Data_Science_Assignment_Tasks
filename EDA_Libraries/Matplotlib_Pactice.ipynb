```python
import numpy as np
import matplotlib.pyplot as plt
```

# Bar Plot

```python
#1. Create a vertical bar chart to compare sales of 5 different products. Add axis labels, title, and customize bar colors.
Product = ["Shoes", "Pants", "Shirts", "Jackets", "Cap"]
Sales = [2200,3334,3244,5345,5523]
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title("Monthly Sales")
plt.bar(Product,Sales, color=['green','blue','red','brown','black'])
plt.show()
```

```python
#2. Plot a horizontal bar chart showing the number of students enrolled in different courses (Python, ML, AI, DSA).
Students = ["Ali", "Bilawal", "Zain", "Usman", "Ammar"]
Course = ['ML','Python','DSA','AI','OOP']
plt.xlabel('Students')
plt.ylabel('Course')
plt.title("Enrollment")
plt.barh(Students,Course)
plt.show()
```

```python
#3. Create a grouped bar chart comparing scores of two students across 5 subjects.
student1=[60,78,50,63,48]
student2=[63,57,75,58,49]
subjects=["Science","English","Math","Urdu","Computer"]
x=np.arange(len(subjects))
width=0.4
plt.title("Student1 VS Student2")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.bar(x-width/2,student1,width,label="student1")
plt.bar(x+width/2,student2,width,label="student2")
plt.xticks(x,subjects)
plt.legend()
plt.show()
```

```python
#4.	Plot a stacked bar chart showing monthly expenses split into categories: rent, food, transport, and savings.
categories=["rent","food","transport","savings"]
June=[1000,2000,3000,400]
July=[500,1000,1500,200]
August=[200,400,600,80]
plt.bar(categories,June,label="June")
plt.bar(categories,July,label="July")
plt.bar(categories,August,label="August")
plt.legend()
plt.title("Monthly Expenses")
plt.ylabel("Categories")
plt.xlabel("Expenses")
plt.show()
```

# Histogram

```python
#5. Generate a histogram showing the distribution of student ages (randomly generated between 18–30). Use different bins values and compare results.
students = ['']
age = [20,25,15,16,22]

plt.hist(age, bins=5, color="blue", edgecolor="black")
plt.title("Distribution of Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#6. Plot a histogram of normally distributed data (mean=50, std=10) and highlight the mean using a vertical line.
data = np.random.normal(50, 10, 1000)
plt.hist(data, bins=15, color="blue", edgecolor="black")
plt.axvline(data.mean(), color="g", label="Mean")
plt.title("Normal Distribution Hist")
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#7. Compare two distributions (e.g., math scores vs science scores) in a single histogram with transparency (alpha).
math = [50,60,76,71,80]
science = [60,70,44,70,80]
plt.title("Comparing score of students")
plt.xlabel("score")
plt.ylabel("frequency")
plt.hist(math,bins=5,alpha=0.4,label="math score",color="green",edgecolor="black")
plt.hist(science,bins=5,alpha=0.4,label="math score",color="blue",edgecolor="black")

plt.show()
```

# Pie Plot

```python
#8. Create a pie chart showing market share of 4 tech companies (Apple, Google, Amazon, Microsoft). Add percentage labels.
companies = ['Apple', 'Google', 'Amazon', 'Microsoft']
shares = [30,30,10,30]

plt.pie(shares,labels=companies, autopct='%1.1f%%')

plt.show()
```

```python
#9. Make a pie chart showing daily activities (work, sleep, exercise, leisure). Explode one slice to emphasize the maximum time spent activity.
activities = ['work', 'sleep', 'exercise', 'leisure']
pert = [40,20,20,20]

ex = [0.2, 0.0, 0.0, 0.0]

plt.pie(pert,labels=activities, autopct='%1.1f%%', explode= ex)

plt.show()
```

```python
#10. Create a donut chart (pie chart with a hole) showing the proportion of machine learning model usage (Regression, Classification, Clustering, Others).
models = ['Regression', 'Classification', 'Clustering', 'Others']
usage = [30, 40, 20, 10]
plt.pie(usage, labels=models, autopct='%1.1f%%',  wedgeprops={'width':0.7})

plt.show()
```

# Scatter Plot

```python
# 11. Create a scatter plot showing relationship between study hours and exam scores. Add labels and a title.
hours = [4,5,6,3,2]
scores = [70,80,90,40,33]

plt.scatter(hours,scores)
plt.title('Score and Study hour Relationship')
plt.xlabel('hours')
plt.ylabel('Scores')

plt.show()
```

```python
#12. Generate a scatter plot of height vs weight, where marker size depends on age and color depends on gender.
height = [5.2,5.4,6,5.0,5.7]
weight = [50,40,75,55,63]
age = [20,24,25,28,27]
gender = ['M','F','M','F','M']

colors = ['blue' if g == 'M' else 'pink' for g in gender]

plt.scatter(height,weight, c= colors, s= age)
plt.title('Height and Weight Relationship')
plt.xlabel('height')
plt.ylabel('weight')
plt.xlim(4,7)
plt.show()
```

```python
#13.Visualize Iris dataset (sepal length vs sepal width) using plt.scatter with different colors for each species.
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data[:, :2]   
y = iris.target       

for species, color in zip([0, 1, 2], ['red', 'green', 'blue']):
    plt.scatter(X[y == species, 0], X[y == species, 1],label=iris.target_names[species], color=color)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
plt.legend()
plt.show()
```

# Customize

```python
#14. Draw a scatter plot with different marker styles (o, x, ^, *) and line styles (--, :, -.) for multiple data series.
x = np.linspace(0, 10, 20)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x)
y4 = np.cos(2*x)

# Plot: marker + line style
plt.plot(x, y1, 'o--', label='sin(x)  (o, --)')   
plt.plot(x, y2, 'x:',  label='cos(x)  (x, :)')    
plt.plot(x, y3, '^-.', label='sin(2x) (^, -.)')   
plt.plot(x, y4, '*-',  label='cos(2x) (*, -)')    

plt.title("Multiple Series — markers and line styles")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
