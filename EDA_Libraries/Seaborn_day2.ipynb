```python
#Q1. Dataset Loading - Load the `tips` dataset using sns.load_dataset("tips"). - Display the first 5 rows and dataset info. 
import seaborn as sns
data = sns.load_dataset('tips')
data.head(5)
```

# Categorical Plot

```python
#Q2. Categorical Plots - Create a boxplot showing total_bill distribution across day. - Create a violinplot showing total_bill across time (Lunch/Dinner). - Draw a stripplot for tip across sex. - Draw a swarmplot for total_bill across smoker. 
sns.boxplot(x = 'total_bill', y= 'day', data=data)
```

```python
sns.violinplot(x='total_bill', y='time', data=data)
```

```python
sns.stripplot(x='tip', y='sex',data=data)
```

# Bar Plots with aggregation

```python
#3. Bar Plots with Aggregation - Create a barplot showing the average tip per day. - Create a barplot showing the average total_bill by sex and smoker. 
sns.barplot(x='day', y='tip',data=data)
```

```python
sns.barplot(x= 'sex', y='total_bill', hue= 'smoker' , data=data)
```

# Pair plot

```python
#Q4. Pair Plots - Use pairplot on the tips dataset to explore relationships among total_bill, tip, and size, colored by sex. 
sns.pairplot(data, hue='sex')
```

# HeatMap

```python
#Q5. Heatmaps - Generate a correlation heatmap of numerical columns (total_bill, tip, size). - Customize the heatmap with annotations and a chosen color palette. 
corr = data.select_dtypes(include=["number"]).corr()
sns.heatmap(corr, annot=True, cmap='plasma')
```

# Customization

```python
#Q6. Customization (palette, style, context) - Apply at least 3 different Seaborn palettes (e.g., "pastel", "deep", "coolwarm") to different plots. - Change the Seaborn style (darkgrid, whitegrid) and context (talk, notebook) to see differences.
sns.barplot(x='sex', y= 'total_bill', data= data )
```
