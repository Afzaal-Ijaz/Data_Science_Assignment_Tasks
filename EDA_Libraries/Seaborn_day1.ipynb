# Scatter Plot

```python
#Q2. Dataset Loading  Load the Iris dataset using sns.load_dataset('iris'). Display the first 5 rows and basic information about the dataset. 
import seaborn as sns
data=sns.load_dataset("iris")
```

```python
print(data.head(5))
```

```python
data.info()
```

```python
#Q3. Distribution Plots (Seaborn Only) Draw a histogram for sepal_length. Draw a KDE plot for sepal_length grouped by species. Draw a distribution plot (histogram + KDE) for petal_length. 
sns.histplot(data['sepal_length'])
```

```python
sns.kdeplot(x="sepal_length",hue='species', data =data)
```

```python
sns.histplot(data['petal_length'],kde=True)
```

# Count Plot

```python
#Q4. Count Plot - Plot the count of each species using Seaborns countplot. \
sns.countplot(data['species'])
```

# Relational Plot

```python
#Q5. Relational Plots Draw a scatter plot showing sepal_length vs sepal_width colored by species. Draw a line plot showing sepal_length vs petal_length grouped by species. 
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=data  )
```

```python
sns.lineplot(x='sepal_length',y='petal_length' ,hue='species',data=data )
```
